# Matplotlib And Seaborn

## Matplotlib
Matplotlib is a tool (or library) in Python that helps you make charts and graphs.
It's like the drawing tool for data — so you can see your numbers as pictures.

- Matplotlib is like the base tool for making graphs in Python.

- It gives you full control over everything — like titles, labels, colors, sizes, and where things go on the chart.

- It’s very powerful, but sometimes you have to write more code to make your chart look nice.

- You can think of it like the engine behind the scenes — other tools (like Seaborn) use it to make prettier plots more easily.

## Seaborn
Seaborn is a Python library that makes it easy to create beautiful charts — like line plots, bar charts, and heatmaps. It’s built on top of Matplotlib, but it does a lot automatically so you don’t have to write as much code.

- It looks better by default — nice colors and clean styles.

- It works really well with pandas DataFrames.

- It helps you make fancy statistical charts (like box plots, violin plots, and correlation heatmaps) with just one line of code.

## Visualization Using Matplotlib and Seaborn

```python
# Importing all required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the file using sns
df = sns.load_dataset('iris')
df
```

#### Histogram

```python
# Making a histogram using matplotlib
# bins=15 sets the number of intervals for the histogram to 15.
# color='green' sets the color of the histogram bars to green.
# alpha=0.5 sets the transparency of the bars to 0.5 (1.0 is fully opaque, 0.0 is fully transparent).

plt.hist(df['sepal_length'], bins = 15, color = 'green', alpha = 0.4 )
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

#### Boxplot

```python
sns.boxplot(data=df, palette ='Set2')
plt.title('Boxplot of Iris Features')
plt.show()
```

```python
# You can make desirfed boxplot by giving it x and y label
sns.boxplot(x='species', y='sepal_length', hue='species', palette ='Set2', data=df)
plt.title('Boxplot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

#### Scatterplot

```python
# Setting the figure size
plt.figure(figsize =(10,5))

# hue='species' colors the points based on the 'species' column, allowing you to distinguish between different species.
# palette='Set1' specifies the color palette to be used for coloring the points based on species.
# s=100 sets the size of the (points) in the scatter plot to 100

sns.scatterplot( data = df, x='sepal_length', y = 'petal_length', hue='species', palette= 'Set2', s=100)
plt.title('Scatterplot of Sepal Length vs. Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

# Show plot
plt.show()
```

#### Line Chart

```python
sns.set_palette('Set2')

plt.plot(df['sepal_length'], label = 'Sepal Length')
plt.plot(df['petal_length'], label = 'Petal Length')
plt.title('Line Chart of Sepal and Petal Length')
plt.xlabel('Index')
plt.ylabel('Length')

# Adding legend for both axis
plt.legend()

# Show Chart
plt.show()
```

#### Pie Chart


```python
# Counting how many times each species appears and storing it in specie_count
specie_count = df['species'].value_counts()

# Setting the palette to set2
sns.set_palette('Set2')

# labels = specie_count.index: these are the names of species
# autopct='%1.1f%%': this shows the percentage on the pie
# startangle=140: this just rotates the pie chart a bit to make it look nicer
plt.pie(specie_count, labels = specie_count.index, autopct= '%1.1f%%', startangle=140, )
plt.title('Species Distribution')

# Show pie chart
plt.show()
```

#### Donut Chart

```python
# Counting how many times each species appears and storing it in specie_count
specie_count = df['species'].value_counts()

# Setting the palette to set2
sns.set_palette('Set2')

# Same as pie chart just adding wedgeprops for donut chart
plt.pie(specie_count, labels = specie_count.index, autopct= '%1.1f%%', startangle= 140, wedgeprops= {'width' : 0.3})
plt.title('Species Distribution')

# Show pie chart
plt.show()
```

#### PairPlot
A pair plot is a special kind of chart that shows all the relationships between multiple columns in your data — all in one big grid of small plots.
- It makes scatter plots for every pair of numeric columns (like height vs weight, weight vs age, etc.)

- It also shows the distribution (like a histogram) of each column on the diagonal.

```python
sns.pairplot(df, hue='species', palette = 'Set2')

# Add a title
plt.suptitle("Pair Plot of Iris Dataset")

# show pair plot
plt.show()
```

```python
sns.pairplot(
    df,
    hue='species',
    palette='Set2',
    markers=["o", "s", "D"],    # Different marker shapes for each species
    plot_kws={'alpha': 0.7}     # Making points slightly transparent
)

# y=1.0 means the title is right above the plot
# y=1.02 means move it a bit higher (2% higher)
# The y value controls the vertical position of the title
plt.suptitle("Pair Plot of Iris Dataset", y=1.02, fontsize=13)

# Show Plot
plt.show()
```
